```python
data.CarName = data.CarName.apply(lambda x: x.split(' ')[0].lower())
data.CarName
```

```python
X = data.drop(['symboling','doornumber','enginelocation','wheelbase','carwidth','carheight','enginetype','fuelsystem','boreratio','stroke','compressionratio','peakrpm','price'], axis = 1)
y = data.price
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape,y.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in X.select_dtypes(include="object").columns.tolist():
    X[i] = le.fit_transform(X[i])
X.head()
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=2529)
```

```python
print('X_train shape:',X_train.shape)
print('X_test shape :',X_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape :',y_test.shape)
```

```python
X.columns
```

```python
var_to_stand=['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel',
       'carlength', 'curbweight', 'cylindernumber', 'enginesize', 'horsepower',
       'citympg', 'highwaympg']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train[var_to_stand]=ss.fit_transform(X_train[var_to_stand])
X_test[var_to_stand]=ss.transform(X_test[var_to_stand])
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.intercept_
```

```python
model1.coef_
```

```python
y1_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_absolute_error,mean_squared_error,r2_score
err1=mean_absolute_percentage_error(y_test,y1_pred)*100
print('mean_absolute_percentage_error:',mean_absolute_percentage_error(y_test,y1_pred))
print('mean_absolute_error:',mean_absolute_error(y_test,y1_pred))
print('mean_squared_error:',mean_squared_error(y_test,y1_pred))
print('r2_score:',r2_score(y_test,y1_pred))
r2_scr1=r2_score(y_test,y1_pred)
```

```python
from sklearn.linear_model import Ridge
model2=Ridge()
```

```python
model2.fit(X_train,y_train)
```

```python
y2_pred=model2.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_absolute_error,mean_squared_error,r2_score
err2=mean_absolute_percentage_error(y_test,y2_pred)*100
print('mean_absolute_percentage_error:',mean_absolute_percentage_error(y_test,y2_pred))
print('mean_absolute_error:',mean_absolute_error(y_test,y2_pred))
print('mean_squared_error:',mean_squared_error(y_test,y2_pred))
print('r2_score:',r2_score(y_test,y2_pred))
r2_scr2=r2_score(y_test,y2_pred)
```
